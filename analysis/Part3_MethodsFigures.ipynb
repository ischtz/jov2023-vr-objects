# Remembered Object Position Task - Analysis - Part 3 - Methods Figure

```python
%matplotlib inline 

import os
import matplotlib.pyplot as plt
from palettable.colorbrewer.qualitative import Set1_9, Set2_8, Paired_12

# All images are square, 1200x1200 px
IMGS = ['../images/example_scene.png', 
        '../images/example_object.png', 
        '../images/example_cube.png', 
        '../images/example_fixed.png', 
        '../images/example_laser.png']
img = [plt.imread(f) for f in IMGS]

TITLES = ['A) Example Scene', 'B) Object Placement', 'C) Cube Placement', 'D) Fixed Pointer', 'E) Laser Pointer']

# Use consistent color scheme for response methods
METHOD_COLORS = [(0,0,0), Paired_12.mpl_colors[1], Paired_12.mpl_colors[7], Paired_12.mpl_colors[3],  Paired_12.mpl_colors[5]]

folder_fig = '../figures'
folder_emm = '../results' # Estimated marginal means data
```

## ðŸ“Š Fig. 1 - Task & Response Methods

```python
# Figure 1: Example scene and response methods
fig, ax = plt.subplot_mosaic([[0, 0, 1, 2],
                              [0, 0, 3, 4]], 
                             dpi=300, layout='constrained', figsize=(7.6, 3.7))

OUTLINE_WIDTH = 1.8
LABEL_SIZE = 10

for ix, image in enumerate(img):
    
    ax[ix].imshow(image, aspect='equal')
    if ix > 0:
        ax[ix].text(0.03, 0.9, TITLES[ix], transform = ax[ix].transAxes, horizontalalignment='left', color='k', fontsize=LABEL_SIZE)
    else:
        ax[ix].text(0.015, 0.95, TITLES[ix], transform = ax[ix].transAxes, horizontalalignment='left', color='k', fontsize=LABEL_SIZE)
    ax[ix].grid(False)

    # Outline color -> indicator method
    for axis in ['top','bottom','left','right']:
        ax[ix].spines[axis].set_linewidth(OUTLINE_WIDTH)
        ax[ix].spines[axis].set_color(METHOD_COLORS[ix])

for axs in ax:
    ax[axs].set_xticklabels([])
    ax[axs].set_xticks([])
    ax[axs].set_yticks([])
    ax[axs].set_yticklabels([])

plt.savefig(os.path.join(folder_fig, 'Fig1_ResponseMethods.png'), dpi=300)
```
